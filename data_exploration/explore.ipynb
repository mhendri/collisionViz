```python
intersection_count = defaultdict(int)
for index, collision in collisions.iterrows():
    intersection_count[(collision['ON STREET NAME'], collision['CROSS STREET NAME'])] += 1
```

```python
sorted_intersection = sorted(intersection_count.items(), key=operator.itemgetter(1), reverse=True )
# print(sorted_intersection)
```

```python
factor1_count = defaultdict(int)
for index, collision in collisions.iterrows():
    factor1_count[(collision['CONTRIBUTING FACTOR VEHICLE 1'], collision['CONTRIBUTING FACTOR VEHICLE 2'])] += 1
```

```python
sorted_factor = sorted(factor1_count.items(), key=operator.itemgetter(1), reverse=True)
print(sorted_factor)
```
